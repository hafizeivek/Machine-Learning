```python
# 5. Hyperparameter Optimization
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# KNN modelini oluştur
knn_model = KNeighborsClassifier()

# Modelin parametrelerini almak
knn_params = {"n_neighbors": range(2, 50)}

# GridSearchCV ile modelin en iyi parametrelerini bul
knn_gs_best = GridSearchCV(knn_model,
                           knn_params,
                           cv=5,                # Cross-validation
                           n_jobs=-1,           # Tüm çekirdekleri kullanarak işlemi hızlandır
                           verbose=1)           # Progresi görsel olarak takip etmek için

# GridSearchCV'yi eğitim verisiyle fit et
knn_gs_best.fit(X, y)

# En iyi parametreyi almak
best_params = knn_gs_best.best_params_

# En iyi parametreyi yazdır
print(best_params)
```
